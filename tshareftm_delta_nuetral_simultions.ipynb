# Total Model Simulation
This simulation takes into account position size, cost of rebalce fees, the approximate apr and is run over 10,000 synthetic histories.

# Strategy Issue
Daily rebalancing ensures the position stays close to delta nuetral however is also significantly effects the CAGR of the strategy. 
For example the no leverage (1x) 80% apr strategy with rebalancing reduces the ending wealth to median of 140% initial while a simple compounding strategy would have resulted in a 220% ending wealth.

There needs to be a strategic rebalancing approach such that the price can be allowed to move and hopefully reach back to 1 and to be rebalanced more infrequently as it approaches some threshold.

### Rebalance at threshold 20%
Rebalance everytime the accum value of the underlying assets reaches +-20%

## Simulating Threshold Rebalancing Strategy

# Leveraged Threshold Rebalancing
The pair TSHARE-FTM is farmed using a leveraged LP position but the first 1x borrowed is TSHARE only to make net nuetral exposure.

Parameters of this model include: size of position in USD ; Collatoral ratio; The total gas costs in USD to rebalance; the average per leverage APR for LP and the leverage used; and the threshold % change in price that must be reached to rebalance position.

This does not take into account rewards are paid in TSHARE and so APR is a function of price.

```python
# initialize parameters
position_size = 10000
#col_ratio = 1
size = 1 #size = 0.5 + 0.5/col_ratio
lp_size = position_size
rebalance_fee = 1.50
fee_dapr = 1.4/365 # 150% APR
leverage = 2.0
threshold = 0.10 # % change in price in either direction to rebalance pos, a log change in price should probably be used in future implimentations

sample = pd.DataFrame(bootstraprets.iloc[:,:])+1
acc_sample = pd.DataFrame(bootstraprets.iloc[:,:])+1
position_val = pd.DataFrame(bootstraprets.iloc[:,:])+1
daily_lp = pd.DataFrame(fee_dapr+sample**0.5)

# the simulated position value using pos size and fees
# interate and cumprod until reach above threshold
for sim in sample.columns:
    current_pos = position_size
    
    for day in sample.index:
        if np.abs(acc_sample.iloc[day,sim]-1) < threshold: 
            '''If accumulating product is below threshold then you keep accumulating the returns.
            Position value becomes the previous day value multiplied by the LP value'''
            lp_value = daily_lp.iloc[day,sim] # value of lev lp
            daily_debt = acc_sample.iloc[day,sim] + acc_sample.iloc[day,sim]*0.5*(leverage-2) + 0.5*(leverage-2) # debt from leverage pos
            daily_comb = lp_value*leverage - daily_debt
            position_val.iloc[day,sim] = daily_comb*current_pos
            if day != 119:
                acc_sample.iloc[day+1,sim]= acc_sample.iloc[day+1,sim]*acc_sample.iloc[day,sim] # keep track of returns
                daily_lp.iloc[day+1,sim]= daily_lp.iloc[day+1,sim]*daily_lp.iloc[day,sim] # keep track of accumulating lp value w/apr
        else:
            '''When the threshold is above you rebalance the position and reset the counter.'''
            # acc_sample.iloc[day] the accumulated return is started again as
            lp_value = daily_lp.iloc[day,sim] # value of lev lp
            daily_debt = acc_sample.iloc[day,sim] + acc_sample.iloc[day,sim]*0.5*(leverage-2) + 0.5*(leverage-2) # debt from leverage pos
            daily_trade = np.abs((acc_sample.iloc[day,sim]-1)*0.5*(leverage)*0.004)  # trading fee to rebalance
            daily_comb = lp_value*leverage - daily_debt - np.abs(daily_trade)
            position_val.iloc[day,sim] = daily_comb*current_pos - rebalance_fee # current position value including gas feesto rebalance
            current_pos = position_val.iloc[day,sim] # after rebalance updatecurrent position value
position_val=position_val/position_size
```

```python
position_val.iloc[:,0:10].plot()
print('$',position_size,' position and ',threshold*100,'% thresh')
print('$',rebalance_fee,' rebalance fee and 0.4% trading fee')
print('APR = ',int(fee_dapr*36500),'%')
print(leverage,'x leverage')
ending_wealth = position_val.iloc[-1,:] # look at the ending wealth of strategy
average_return = ending_wealth.mean()
median_return = ending_wealth.median()
fifth_quantile = np.percentile(ending_wealth, .05)
cagr = median_return**(1/120)
print('Average return: ',average_return)
print('Median return: ',median_return)
print('5th quantile return: ',fifth_quantile)
print('CAGR: ',cagr, (cagr-1)*365)
```

# Impact of different fee values and positions

```python
# initialize parameters
position_size = 2500
col_ratio = 0.625
size = 0.5 + 0.5/col_ratio
lp_size = position_size*(1/size)
short_size = position_size*((size-1)/size)
rebalance_fee = 1.75
fee_dapr = 0.8/365 # 150% APR
leverage = 5.0
threshold = 0.1 # % change in price in either direction to rebalance pos, a log change in price should probably be used in future implimentations

sample = pd.DataFrame(bootstraprets.iloc[:,:])+1
acc_sample = pd.DataFrame(bootstraprets.iloc[:,:])+1
position_val = pd.DataFrame(bootstraprets.iloc[:,:])+1
daily_lp = pd.DataFrame(fee_dapr+sample**0.5)

# the simulated position value using pos size and fees
# interate and cumprod until reach above threshold
for sim in sample.columns:
    current_pos = position_size
    
    for day in sample.index:
        if np.abs(acc_sample.iloc[day,sim]-1) < threshold: 
            '''If accumulating product is below threshold then you keep accumulating the returns.
            Position value becomes the previous day value multiplied by the LP value'''
            lp_value = daily_lp.iloc[day,sim] # value of lev lp
            daily_short = 1/col_ratio - acc_sample.iloc[day,sim] # value of short / value of collatoral - borrowed
            daily_debt = acc_sample.iloc[day,sim]*0.5*(leverage-1) + 0.5*(leverage-1) # debt from leverage pos
            daily_comb = lp_value*leverage - daily_debt
            position_val.iloc[day,sim] = (daily_comb + daily_short*0.5)*current_pos*(1/size)
            if day != 364:
                acc_sample.iloc[day+1,sim]= acc_sample.iloc[day+1,sim]*acc_sample.iloc[day,sim] # keep track of returns
                daily_lp.iloc[day+1,sim]= daily_lp.iloc[day+1,sim]*daily_lp.iloc[day,sim] # keep track of accumulating lp value w/apr
        else:
            '''When the threshold is above you rebalance the position and reset the counter.'''
            # acc_sample.iloc[day] the accumulated return is started again as
            lp_value = daily_lp.iloc[day,sim] # value of lev lp
            daily_short = 1/col_ratio - acc_sample.iloc[day,sim] # value of short
            daily_debt = acc_sample.iloc[day,sim]*0.5*(leverage-1) + 0.5*(leverage-1) # debt from leverage pos
            daily_trade = np.abs((acc_sample.iloc[day,sim]-1)*0.5*(leverage)*0.005)  # trading fee to rebalance
            daily_comb = lp_value*leverage - daily_debt - np.abs(daily_trade)
            position_val.iloc[day,sim] = (daily_comb + daily_short*0.5)*current_pos*(1/size) - rebalance_fee # current position value including gas feesto rebalance
            current_pos = position_val.iloc[day,sim] # after rebalance updatecurrent position value
position_val=position_val/position_size
```
